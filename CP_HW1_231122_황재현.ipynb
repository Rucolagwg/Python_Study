```python
import numpy as np
import matplotlib.pyplot as plt

# 1번
def Taylor_log(x, N):
    add = 0
    for n in range(1, N): # N-1차까지
        add += -(-1)**n * (x**n)/n # 테일러 급수의 합
    return add

# 2번
X = np.linspace(0, 0.9, 19)

# 3번
fig = plt.figure()
ax = fig.add_subplot()
# Y1 = np.log(1+X)

ax.plot(X, np.log(1+X), label = 'original')

for i in [3, 5, 7]: # 아래 코드를 for문으로 나타낸 것
    ax.scatter(X, Taylor_log(X, i), label = '$N_{max}$' + f' = {i-1}') # 중간 내용은 시험에 내는건 아니야
#ax.scatter(X, Taylor_log(x, 3), label = 'Nmax = 2')
#ax.scatter(X, Taylor_log(x, 5), label = 'Nmax = 4')
#ax.scatter(X, Taylor_log(x, 7), label = 'Nmax = 6')
ax.set_xlabel("x")
ax.set_ylabel("log(1+x)")
ax.legend()
plt.show()
```

**4번**

```python
def compute_avg_error(x, N):
    # np.log(1+x) - Taylor_log(x, N)의 절댓값
    error = np.abs(np.log(1+x) - Taylor_log(x, N))
    # 평균 오차 계산
    avg_error = np.mean(error) #np.mean은 평균을 구하는 모듈
    return avg_error

# 테스트
N = 7
avg_error = compute_avg_error(X, N)
print("평균 오차:", avg_error)
```

**5번**

```python
N = 1
avg_error = compute_avg_error(X, N)

# 평균 오차가 1e-3 아래로 줄어드는 N 값 찾기
while avg_error > 1e-3: # 평균 오차가 목표 오차보다 큰 동안에는 N에 1을 더해야한다.
    N += 1
    avg_error = compute_avg_error(X, N)

print("평균 오차가 1e-3 아래로 줄어드는 N 값:", N)
```

```python
def Taylor_log(x, N):
    sum = 0.0
    for n in range(1, N):
        sum += ((-1) ** (n - 1)) * (x ** n) / n

    return sum

x = 1
N_val = range(1, 21) # 테일러 급수의 변화를 관찰하기 위해 N=1부터 20까지의 범위 설정
y_val = [Taylor_log(x, N) for N in N_val]

plt.plot(N_val, y_val, marker='o')
plt.xlabel('$N$')
plt.ylabel('$log(2)$')
plt.show()
```

```python
def plot(dx):
    N = range(1, 40)

    fig = plt.figure()
    ax = fig.add_subplot()

    ax.hlines(np.log(1+dx), N[0], N[-1])

    Ta = [ Taylor_log(dx, n) for n in N ]
    ax.scatter(N, Ta)

    ax.set_xlabel("$N$")
    ax.set_ylabel(f"Taylor_log({dx}, N)")
    plt.show()

plot(1.0)
plot(1.1)
```
