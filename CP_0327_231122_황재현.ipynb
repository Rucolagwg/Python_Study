```python
import numpy as np
import matplotlib.pyplot as plt

def plot_func(x, y, title=''):
    fig = plt.figure(figsize=(6,4))
    ax = fig.add_subplot() #Make an axis

    ax.plot(x, y)
    ax.scatter(x[::10], y[::10], marker='x') #Use Slicing
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    plt.show()

x1 = np.linspace(0, 25, 101)
y1 = np.sin(x/np.pi)
plot_func(x1, y1, title = '$y=x^2$')

x2 = np.linspace(-3, 3, 101)
y2 = np.exp(x)
plot_func(x2, y2, title = '$y=exp(x)$')

x3 = np.linspace(0, 25, 101)
y3 = x**3
plot_func(x3, y3, title = '$y=x^3$')

x4 = np.linspace(0, 25, 101)
y4 = np.log(x+1)
plot_func(x4, y4, title = '$y=\log(x+1)$')

x5 = np.linspace(0, 25, 101)
y5 = x + 2 * (x ** 2)
plot_func(x5, y5, title = '$y=x+2x^2$')
```
