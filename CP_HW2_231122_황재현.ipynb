```python
import numpy as np

fname = path + '/MyDrive/data/pn.txt'

overall = np.genfromtxt(fname)
overall.shape
```

2.1

```python
delta_x = overall[1, 0] - overall[0, 0]

sum = np.sum(overall[:, 1] * overall[:, 2] * delta_x)

print("P0(x)P1(x)Δx의 합:", sum)
```

2.2

```python
def sum_pp(overall, n1, n2):
    return np.sum(overall[:, n1 + 1] * overall[:, n2 + 1] * delta_x)

# n1 = 0, n2 = 1
print("P0(x)P1(x)Δx의 합:", sum_pp(overall, 0, 1))
```

2.3

```python
for n1 in range(5):
    for n2 in range(5):
        if n1 != n2:
            result = sum_pp(overall, n1, n2)
            print(f"P{n1}(x)P{n2}(x)Δx의 합:", result)
```

```python
import matplotlib.pyplot as plt

x_values = overall[:, 0]

def plot_func(n_values, scatter=False):
    for n in n_values:
        y_values = overall[:, n + 1]

        if scatter:
            plt.plot(x_values, y_values, label=f'P{n}(x)')
            plt.scatter(x_values[::5], y_values[::5])
        else:
            plt.plot(x_values, y_values, label=f'P{n}(x)')

    plt.xlabel('$x$')
    plt.ylabel('$P_n(x)$')
    plt.legend()
    plt.show()

n_values = [0, 1, 2, 3, 4]
plot_func(n_values, scatter=False)
plot_func(n_values, scatter=True)
```
