20번

```python
def calculate(n, start = 1):
    sum = 0
    for i in range(start, n+1):
        sum += i
    return sum

n = 7
m = 4
result = calculate(n, start = m)
print(f"{m}부터 {n}까지의 합은 {result}")
```

21번

```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

n = 4
print(factorial(n))
```

```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

def fac_factorial(n): # n!의 factorial을 구하기 위함
    return(factorial(factorial(n)))

n = 3
print(fac_factorial(n))
```

22번

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_func(x, title = None):
    y1 = x**2
    y2 = np.sin(x)

    plt.plot(x, y1, label = '$y = x^2$')
    plt.plot(x, y2, label = '$y = sin(x)$')
    plt.xlabel('x')
    plt.ylabel('y')

    if title is not None:
        plt.title(title)

    plt.show()

x = np.linspace(-10, 10, 101)
plot_func(x, title = "$y=x^2$, $y=sin(x)$")
```
