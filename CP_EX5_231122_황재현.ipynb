```python
import matplotlib.pyplot as plt

fig = plt.figure(131)

ax1 = fig.add_subplot()
for n, y in enumerate(Ys):
    if n == 0 or n == 1 or n == 2:
        ax1.plot(x, y, label=f'$n$={n}')

ax1.set_xlabel("$x$")
ax1.set_ylabel("$J_n(x)$")
ax1.set_title("Bessel functions of the first kind")
ax1.legend()

plt.show()
#------------------------------------------------------
fig = plt.figure(132)

ax2 = fig.add_subplot()
for n, y in enumerate(Ys):
    if n == 1 or n == 3 or n == 5:
        ax2.plot(x, y, label=f'$n$={n}')

ax2.set_xlabel("$x$")
ax2.set_ylabel("$J_n(x)$")
ax2.set_title("Bessel functions of the first kind")
ax2.legend()

plt.show()
#------------------------------------------------------
fig = plt.figure(133)

ax3 = fig.add_subplot()
for n, y in enumerate(Ys):
    if n % 3 == 0:
         ax3.plot(x, y, label=f'$n$={n}')
         ax3.scatter(x[::100], y[::100])

ax3.set_xlabel("$x$")
ax3.set_ylabel("$J_n(x)$")
ax3.set_title("Bessel functions of the first kind (with scatter)")
ax3.legend()

plt.show()
```
