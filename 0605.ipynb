Midpoint

```python
import numpy as np

def f(x):
    return x*x

N= 20 # N이 증가할수록 값이 정확해진다.
X = np.linspace(0, 1, N) # 정확히 맞아들어가지 않기에 숫자가 0.33333 이렇게 출력되지 않음
sum = 0
for x1, x2 in zip(X[:-1], X[1:]):
    sum += f((x1+x2)/2)*(x2-x1)

sum
```

Trapezoidal method

```python
def f(x):
    return x*x

N = 20
X = np.linspace(0, 1, N)
sum = 0
for x1, x2 in zip(X[:-1], X[1:]):
    sum += ( f(x1) + f(x2) ) * (x2-x1) / 2

sum
```

```python
def f(x):
    return x*x

N = 20
X = np.linspace(0, 1, N)
sum = 0
for x1, x2, x3 in zip(X[::2], X[1::2], X[2::2]):
    sum += ( f(x1) + 4 * f(x2) * f(x3)) * (x2-x1) / 2

sum
```

```python
from scipy import integrate

def ylower(x):
    return 0

def yupper(x):
    return 1-x

def func(y, x):
    return x*y

integrate.dblquad(func, 0, 1, ylower, yupper)
```
