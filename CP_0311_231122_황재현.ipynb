```python
sum = 0
seq = []
a = 1000
while a < 2001:
    if a % 3 == 0:
        sum += a
        a += 1
        seq.append(sum)
    else:
        a += 1
print(seq[-1])
```

```python
a = 1
b = 1
n = 0
seq = [1, 1] #a_1, a_2
while n < 10:
    c = b + a
    a = b
    b = c
    n += 1
    seq.append(c) #a_2부터 저장
print(seq)
```
