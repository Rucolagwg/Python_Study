```python
def Ef_origin(rx, ry, q):
    deno = (rx*rx + ry*ry) ** (3/2)
    return (rx*q / deno, ry*q / deno)

print(Ef_origin(1, 0, 1))
print(Ef_origin(2, 0, 1))
print(Ef_origin(1, 0, -1))
print(Ef_origin(1, 1, 1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Efield_2D(rx, ry, q = [1], q_pos = [(0,0)]):
    shape = [2] + list(rx.shape)
    Ef = np.zeros(shape)
    for qi, (x, y) in zip(q, q_pos):
        Ef += np.array ([rx-x, ry-y]) * qi / ((rx-x)**2 + (ry-y)**2) ** (3/2)
    return Ef

fig = plt.figure()
ax = fig.add_subplot(aspect= 'equal')

Nplot = 100
X = np.linspace(-2, 2, Nplot)
Y = np.linspace(-2, 2, Nplot)

xx, yy = np.meshgrid(X, Y)
#E = Efield_2D(xx, yy)
#print(E, E.shape)
Ex, Ey = Efield_2D(xx, yy)

ax.streamplot(xx, yy, Ex, Ey)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Electric Field")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pot_2D(rx, ry, q = [1], q_pos = [(0,0)]):
    E = 0
    for qi, (x, y) in zip(q, q_pos):
        E += qi / np.sqrt((rx-x)**2 + (ry-y)**2)
    return E

fig = plt.figure()
ax = fig.add_subplot(aspect = 'equal')

Nplot = 100
X = np.linspace(-2, 2, Nplot)
Y = np.linspace(-2, 2, Nplot)

xx, yy = np.meshgrid(X, Y)
qn = [(-1)**n for n in range(0, 6)]
qn_pos = [(np.cos(2*n*np.pi/6), np.sin(2*n*np.pi/6)) for n in range(0, 6)]
Ex, Ey = Efield_2D(xx, yy, qn, qn_pos)
pot = pot_2D(xx, yy, q = [1, -1, 1, -1, 1, -1], q_pos=qn_pos)
cont = ax.contourf(xx, yy, pot, cmap = 'bwr', levels = 100)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Electric Potential")

plt.show()
```
