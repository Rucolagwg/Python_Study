```python
original = 'a,b,c,d,e,f'
original.split(',')
```

```python
sp = []
a = 0
while a < len(original):
    if original[a] != ',':
        sp += original[a]
    a += 1
sp
```

```python
','.join(sp)
```

```python
'-'.join(sp)
```

```python
sum = ''
for x in sp:
    print(x)
```

```python
sum = ''
for x in sp:
    sum += x
sum
```
