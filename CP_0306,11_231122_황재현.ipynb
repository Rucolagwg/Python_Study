```python
fac = 1
sum = 0
n = 0
while n < 5:
    n += 1
    fac = fac * n # fac *= n
    sum += fac
    print("n = ", n, ", fac = ", fac, ", sum = ", sum)
print("sum =", sum)
```

```python
fac = 1
sum = 0
n = 0
while n < 5:
    n += 1
    fac = fac * (2*n-1)
    sum += fac
    print("n = ", n, ", fac = ", fac, ", sum = ", sum)
print("sum =", sum)
```
