```python
name_vote = path + '/MyDrive/data/votes.txt'

count = 0
n_empty = 0
n_occupied = 0
info = []
points = 0

with open(name_vote, 'r') as f: # with open 구문을 통해 f가 구문 내에서만 효력을 가지도록 하게 함
#f = open(name_vote, 'r')
    for line in f:
        sline = line.split()
        if len(sline) > 0:
            n_occupied += 1
            info += [sline] # 대괄호를 안하면, list끼리 더한거라 일차원 리스트가 만들어진다
            points += int(sline[1])
        else:
            n_empty += 1
            if points != 10:
                print("The sum wasnt 10")
            points = 0
       #print(line, end = '')  # line: 문자열 str / sline: list
        count += 1

print(f'{count} lines, {n_empty} empty lines, {n_occupied} occupied lines')
```
