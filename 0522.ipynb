```python
class exam:
    def __init__(self, label, scores):
        self.name = label
        self.scores = scores
        self.N = len(scores)
        self.avg = self.get_avg() # 속성이 정의됨
        self.var = self.get_var()
        self.std = np.sqrt ( self.var )

    def get_avg(self):
        return np.sum(self.scores) / self.N

    def get_var(self):
        add = 0
        for x in self.scores:
            add += (x - self.avg)**2

        return add/self.N

    def get_var_v2(self):
        return np.sum( (self.scores - self.avg)**2 ) / self.N

    def scatter(self):
        fig = plt.figure()
        ax = fig.add_subplot()
        ax.scatter(range(len(self.scores)), self.scores)

        return ax

    def get_dist(self):
        num = np.zeros(11)
        for x in self.scores:
            num[int(x/10)] += 1

        return num

    def get_dist_gen(self, bins = 5):
        min, max = np.min(self.scores), np.max(self.scores)
        width = (max - min)/bins
        edges = np.array( [min + i * width for i in range(bins+1)] )
        num = np.zeros(bins + 1)
        for x in self.scores:
            num[ int( (x-min) / width) ] += 1
        num[-2] += num[-1]

        return num[:-1], edges

    def plot_hist(self, bins = 5):
        hist, edges = self.get_dist_gen(bins=bins)
        fig = plt.figure()
        ax = fig.add_subplot()
        ax.bar(edges[:-1], hist, label='at left edge')
        centers = [ (a+b)/2 for a, b in zip(edges[0:-1], edges[1:]) ]
        ax.bar(centers, hist, width = (edges[1]-edges[0])*0.8, label = 'at center')
        ax.legend()
        ax.set_title(self.name)

        return ax
```

```python
import numpy as np
import matplotlib.pyplot as plt

scores = np.genfromtxt(path + '/MyDrive/Data/scores.txt')
SA = exam('A', scores[:,0]) # 첫번째 열만 떼어냄
SB = exam('B', scores[:,1])
print(f'{SA.avg}, {SB.avg}')
SA.plot_hist()
```
