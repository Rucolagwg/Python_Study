```python
row0 = ['Red', 10, 'high']
row1 = ['Blue', 15, 'low']
print(row0)
print(row1)
total = [row0, row1]
print(total)
print(total[0]) #total[0]는 row0와 같다.
print(row0[0])
total #중간에 입력하면 출력해서 보여주진 않는다.
print(total[0][1])
total[0] #row0 출력과 같은 것, 리스트의 첫번째 요소니까
```
