```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(1)

def walk(N_toss):
    position = 0
    for _ in range(N_toss):
        if np.random.random() > 0.5:
            position += 1
        else:
            position -= 1
    return position

def random_walk(N_walker, N_toss):
    pos = dict()
    for _ in range(N_walker):
        num = walk(N_toss)
        pos[num] = pos.get(num, 0) + 1
    return pos

result = random_walk(100, 100)

def plot(N_walker, N_toss):
    x = result.keys()
    y = result.values()

    fig = plt.figure()
    ax = fig.add_subplot()
    ax.scatter(x, y)

plot(10000, 10000)
```
