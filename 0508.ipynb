```python
class MyClass:
    num = 1
    def show_num(self):
        print("self.num = ", self.num)
    def add_num(self, add): # self는 있어야 한다. 추가 인자에 대한 정의도 필요
        self. num += add
```

```python
c1 = MyClass()
c1.show_num() # 함수 호출이니 괄호 필요. 여기서 self는 굳이 다시 안써도 된다.
c1.add_num(3)
c1.show_num() # 캡슐화 되어 있다. 반복 작업시 유용
c1.num += 10
c1.num
```

```python
c1 = MyClass()
c2 = MyClass()
c1.show_num()
c1.add_num(3)
c1.show_num()
c2.show_num()
```

```python
# 학번을 Student 인스턴스 생성할 때 외부인자로 넣어줄 수 있게끔
#__init__() method를 Student class에 추가해보기
# 학년 진급 함수

class Human:
    name = ''
    age = 0

class Student(Human):
    def __init__(self, id, name, age):
        self.id = id
        self.name = name
        self.age = age
        self.year = 1
    def show_info(self):
        print('id:', self.id, 'name:', self.name, 'age:', self.age, 'year:', self.year)
    def promote(self):
        self.year += 1
```

```python
s1 = Student(1000, 'Hwang', 21)
s1.show_info() # self는 안쓰는거야
s1.promote()
s1.show_info()
```
